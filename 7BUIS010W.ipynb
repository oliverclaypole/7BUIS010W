```python
import pandas as pd
import yfinance as yf

url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
    # Colab suggestion to add headers to mimic a web browser to avoid 403 Forbidden error due to wikipedia server thinking i'm a bot

tables = pd.read_html(url, storage_options=headers) # extracting tables
SP500_tickers = tables[0] # selecting S&P 500 table
print(SP500_tickers.head())



ticker_list = [ticker.replace('.','-') for ticker in SP500_tickers['Symbol'].tolist()]
  # creates a list of the tickers from the table
  # ticker.replace removing data error with '.'
stock_data = yf.download(ticker_list, start='2022-01-01', end='2025-01-01') # downloading data
stock_prices = stock_data['Adj Close'] # we only require the Adj close for our calculations
stock_prices_new = stock_prices.dropna(axis=5) # removing any company that has one missed value in the data to avoid NaN's
print(f'original: {stock_prices.shape[1]}stocks')
print(f'original: {stock_prices.shape[1]}stocks')

print (stock_prices_new.head())
print(stock_prices_new.shape)
```
